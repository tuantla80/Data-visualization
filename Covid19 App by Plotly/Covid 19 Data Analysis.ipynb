```python
import pandas as pd
```

## 1. Get Covid 19 datasets 
https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_time_series

```python
url_confirmed = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
url_deaths = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
url_recovered = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
```

### Get confirmed cases and un-pivot

```python
confirmed = pd.read_csv(url_confirmed)
confirmed.head()
```

```python
confirmed.to_csv('time_series_covid19_confirmed_global.csv', index=False)
```

```python
date1 = confirmed.columns[4:]
total_confirmed = confirmed.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                 value_vars=date1, 
                                 var_name='date', 
                                 value_name='confirmed')
total_confirmed.head(3)
```

```python
total_confirmed.tail(5)
```

### Get death cases and un-pivot

```python
deaths = pd.read_csv(url_deaths)
```

```python
deaths.to_csv('time_series_covid19_deaths_global.csv', index=False)
```

```python
deaths.head()
```

```python
date2 = deaths.columns[4:]
total_deaths = deaths.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                           value_vars=date2, 
                           var_name='date',
                           value_name='death')
total_deaths.head(3)
```

```python
total_deaths.tail(5)
```

### Get recovered cases and un-pivot

```python
recovered = pd.read_csv(url_recovered)
recovered.head()
```

```python
recovered.to_csv('time_series_covid19_recovered_global.csv', index=False)
```

```python
date3 = recovered.columns[4:]
total_recovered = recovered.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                 value_vars=date3, 
                                 var_name='date', 
                                 value_name='recovered')
total_recovered.head(3)
```

```python
total_recovered.tail(5)
```

## 2. Merging dataframes: total_confirmed, total_deaths, total_recovered

```python
covid_data = total_confirmed.merge(right = total_deaths, 
                                   how = 'left', 
                                   on = ['Province/State', 'Country/Region', 'date', 'Lat', 'Long'])
covid_data.head(3)
```

```python
covid_data.tail(5)
```

```python
covid_data = covid_data.merge(right = total_recovered, 
                              how = 'left', 
                              on = ['Province/State', 'Country/Region', 'date', 'Lat', 'Long'])
covid_data.head(3)
```

```python
covid_data.tail(5)
```

### Change column datatypes

```python
covid_data['date'] = pd.to_datetime(covid_data['date'])
```

```python
covid_data.head(3)
```

```python
covid_data.shape
```

```python
covid_data.isna().sum()
```

### Replace recovered: NaN -> 0

```python
covid_data['recovered'] = covid_data['recovered'].fillna(0)
```

```python
covid_data.isna().sum()
```

## 3. Data aggregation

### Calculate Active cases

```python
covid_data['active'] = covid_data['confirmed'] - covid_data['death'] - covid_data['recovered']
```

```python
covid_data.head(3)
```

```python
covid_data.tail(5)
```

```python
covid_data.shape
```

```python
# Just for testing
covid_data['date'].iloc[-1].strftime('%B %d, %Y')  # iloc[-1]: choose last index
```

### Group by date

```python
# Sum over date of all countries
covid_data_by_date = covid_data.groupby(['date'])[['confirmed', 'death', 'recovered', 'active']].sum().reset_index()
covid_data_by_date
```

#### The last row is the current (today) number cases

```python
covid_data_by_date['confirmed'].iloc[-1] # similarly for death, recovered and active cases
```

#### The last second row is the yesterday number cases

```python
covid_data_by_date['confirmed'].iloc[-2] # similarly for death, recovered and active cases
```

#### New confirmed cases

```python
covid_data_by_date['confirmed'].iloc[-1] - covid_data_by_date['confirmed'].iloc[-2]
```

```python
# percentage of changing from yesterday and today
round(((covid_data_by_date['confirmed'].iloc[-1] - covid_data_by_date['confirmed'].iloc[-2]) / covid_data_by_date['confirmed'].iloc[-1]) * 100, 2)
```

### Group by date and country

```python
covid_data_by_date_country = covid_data.groupby(['date', 'Country/Region'])[['confirmed', 'death', 'recovered', 'active']].sum().reset_index()
covid_data_by_date_country
```

#### Test for Australia

```python
# confirmed cases: today-yesterday in Australia
value_confirmed = covid_data_by_date_country[covid_data_by_date_country['Country/Region'] == 'Australia']['confirmed'].iloc[-1] - covid_data_by_date_country[covid_data_by_date_country['Country/Region'] == 'Australia']['confirmed'].iloc[-2]
value_confirmed
```

```python
covid_data_by_date_Australia = covid_data_by_date_country[covid_data_by_date_country['Country/Region'] == 'Australia'][['Country/Region', 'date', 'confirmed']].reset_index()
covid_data_by_date_Australia
```

```python
covid_data_by_date_Australia['daily confirmed'] = covid_data_by_date_Australia['confirmed'] - covid_data_by_date_Australia['confirmed'].shift(1)
```

```python
covid_data_by_date_Australia
```

#### Moving average

```python
# Moving windows
covid_data_by_date_Australia['Moving Average'] = covid_data_by_date_Australia['daily confirmed'].rolling(window=7).mean()
```

```python
covid_data_by_date_Australia.head(20)
```
